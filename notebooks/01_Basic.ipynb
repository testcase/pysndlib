```python
import pysndlib.clm as clm
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
from matplotlib import pyplot as plt
```

```python
with clm.Sound(play=False):
    osc = clm.make_oscil(300)
    mod = clm.make_oscil(231)
    ienv = clm.make_env([0.0,0.0,1.0,1.0], length=44100)
    for i in range(44100):
        clm.outa(i, clm.oscil(osc, clm.oscil(mod)*.01*clm.env(ienv)))
ipd.Audio("test.wav", autoplay=True)
```

```python
a, sr = clm.file2ndarray('test.wav')
np.shape(a)
```

```python
plt.figure(figsize=(18, 2.5))
plt.plot(a[0], color='gray')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
